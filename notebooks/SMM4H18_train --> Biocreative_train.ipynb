```python
start, end = [], []
data = []
for i,(tid, uid, cid, s,l, dn, dnm) in enumerate(zip(smmdf['tweet_id'], smmdf['user_id'], 
                                                     smmdf['created_at'], smmdf['text'], 
                                                     smmdf['label'], smmdf['Drug Name Extracted'], 
                                                     smmdf['Drug Name Normalized'])):
    
    st, en = '-', '-'
    
    if l == 1:
        #print(i,s.lower(),l)
        #print('looking for drug:', dn.split('; '))
        #for i,ss in enumerate(s):
         #   print(i,ss)
        print(dn,'|',dnm)
        try:
            dnm.split('; ')
        except:
            dnm = dn
        for d, dd in zip(dn.split('; '), dnm.split('; ')):
            for match in finditer(d.lower().strip(), s.lower()):
                #print(match.span(), match.group())
                data.append((tid, uid, cid, s, *match.span(), match.group(),dd))
        #print()
    else:
        data.append((tid, uid, cid, s, '-', '-','-', '-'))
```
